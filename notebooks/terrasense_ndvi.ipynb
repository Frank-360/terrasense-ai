```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated satellite bands
red = np.random.rand(100,100)
nir = np.random.rand(100,100)

# NDVI calculation
ndvi = (nir - red) / (nir + red)

plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title("TerraSense NDVI Map")
plt.show()
```
